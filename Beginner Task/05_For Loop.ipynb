```python
# 1. Program to simulate rolling a six-sided die at least 20 times

import random
rolls = 20
count_6 = 0
count_1 = 0
two_sixes_in_row = 0
previous_roll = 0 

# Roll the die 20 times using for loop
for i in range(rolls):
    roll = random.randint(1, 6)  
    print("Roll", i + 1, ":", roll)

    if roll == 6:
        count_6 += 1

        if previous_roll == 6:
            two_sixes_in_row += 1

    if roll == 1:
        count_1 += 1

    previous_roll = roll

print("\nStatistics:")
print("Number of times rolled 6:", count_6)
print("Number of times rolled 1:", count_1)
print("Number of times rolled two 6s in a row:", two_sixes_in_row)
```

```python
# 2. Program to simulate a jumping jacks workout routine

total_jumping_jacks = 100
completed = 0

# Perform jumping jacks in sets of 10
for i in range(10):
    completed += 10  
    remaining = total_jumping_jacks - completed

    print("You completed", completed, "jumping jacks.")

    tired = input("Are you tired? (yes/y or no/n): ").lower()

    # If user says yes
    if tired == "yes" or tired == "y":
        skip = input("Do you want to skip the remaining sets? (yes/y or no/n): ").lower()

        if skip == "yes" or skip == "y":
            print("You completed a total of", completed, "jumping jacks.")
            break

    # If user says no
    if tired == "no" or tired == "n":
        if remaining > 0:
            print(remaining, "jumping jacks remaining.\n")

    # If all jumping jacks are completed
    if completed == total_jumping_jacks:
        print("Congratulations! You completed the workout.")
        break
```
