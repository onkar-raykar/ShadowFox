```python
# 1. Write a function that takes two arguments, 145 and o, returns formatted string. Print and identify representation.
def format_number(num, ch):
    return "Number: {}, Character: {}".format(num, ch)

result = format_number(145, 'o')
print(result)
```

```python
# 2. Calculate area of circle: Area = πr² .
pi = 3.14
radius = 84

# Area of circle
area = pi * radius * radius
print("Pond Area:", int(area))

# Total amount of water in the pond
water_per_sq_meter = 1.4
total_water = area * water_per_sq_meter
print("Total Water in Pond:", int(total_water))
```

```python
# 3. Cross 490m long street in 7 minutes, speed in m/s. Print without decimal.
distance = 490       # meters
time_minutes = 7
time_seconds = time_minutes * 60

speed = distance / time_seconds
print("Speed (m/s):", int(speed))
```
