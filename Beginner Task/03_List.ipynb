# 03 List

```python
# List of Justice League members
justice_league = ["Superman", "Batman", "Wonder Woman", "Flash", "Aquaman", "Green Lantern"]
print("Initial Justice League:", justice_league)

# 1. Calculate the number of members in the Justice League
print("\n1. Number of members:", len(justice_league))
```

```python
# 2. Batman recruited Batgirl and Nightwing as new members. Add them to the list.
justice_league.append("Batgirl")
justice_league.append("Nightwing")
print("\n2. After adding Batgirl and Nightwing:", justice_league)
```

```python
# 3. Wonder Woman is now the leader of the Justice League. Move her to the beginning.
justice_league.remove("Wonder Woman")
justice_league.insert(0, "Wonder Woman")
print("\n3. After making Wonder Woman the leader:", justice_league)
```

```python
# 4. Separate Aquaman and Flash by placing Superman in between them
justice_league.remove("Superman")
flash_index = justice_league.index("Flash")
justice_league.insert(flash_index, "Superman")
print("\n4. After separating Aquaman and Flash:", justice_league)
```

```python
# 5. Replace the existing list with a new team
justice_league = ["Cyborg", "Shazam", "Hawkgirl", "Martian Manhunter", "Green Arrow"]
print("\n5. New Justice League team:", justice_league)
```

```python
# 6. Sort the Justice League alphabetically and select new leader at index 0
justice_league.sort()
print("\n6. Sorted Justice League:", justice_league)
print("New Leader:", justice_league[0])
```
